```python
# 데이터 불러오기
root_dir = os.getcwd()
f_lists = os.listdir(root_dir)
print("File Lists : ", f_lists)
```

root_dir로 지정한 디렉토리에서 listdir 함수를 통해 해당 디렉토리
내 모든 파일 리스트를 가져온다.

```python
# csv파일로 정제된 파일만 불러오기
data_list = [f for f in f_lists if f.endswith('.csv')]
print("Data Lists : ", data_list)
```

모든 파일 리스트가 담긴 리스트에서 요소를 하나 꺼내와서 f로 명명하며, f가 ‘.csv’
로 끝나면 새로운 리스트로 구성하고 불러온다.

```python
# 데이터를 가져오기 위한 함수를 정의
def csv_read_(data_dir, data_list):
    tmp = pd.read_csv(os.path.join(data_dir, data_list), sep=',', encoding='UTF8')
    y, m, d =map(int, data_list.split('-')[-1].split('.')[:-1])
    time = tmp['Time']
    tmp['DTime'] ='-'.join(data_list.split('-')[-1].split('.')[:-1])
    ctime = time.apply(lambda _ : _.replace(u'오후', 'PM').replace(u'오전', 'AM'))
    n_time = ctime.apply(lambda _ : datetime.datetime.strptime(_, "%p %I:%M:%S"))
    newtime = n_time.apply(lambda _ : _.replace(year=y, month=m, day=d))
    tmp['Time'] = newtime
    return tmp
```

ctime은 Time 값의 오전은 AM으로 오후는 PM으로 바꾸고,

n_time은 오전/오후, 시, 분, 초로 바꾸고,

newtime은 년도-월-일로 바꿔 출력한다.

```python
# 정의한 함수를 이용하여 모든 파일의 데이터를 합쳐서 불러옴
dd = csv_read_(root_dir, data_list[0])
for i in range(1, len(data_list)):
    dd = pd.merge(dd, csv_read_(root_dir, data_list[i]), how='outer')
dd
```

```python
# drop 함수를 이용해 사용하지 않는 index 컬럼 제거
dd = dd.drop('Index', axis=1)
dd
```

```python
# set_index 함수를 통해 Time 컬럼을 인덱스로 설정
dd = dd.set_index('Time')
dd
```
